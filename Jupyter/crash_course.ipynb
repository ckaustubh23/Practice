### String Formatting

```python
#defaut order
string = "{} {} {}".format("Python", "Progrmming", "Language")
string
```

```python
# index based
string = "{1} {0} {2}".format("Python", "Progrmming", "Language")
string
```

```python
# keywrd based
string = "{c} {a} {b}".format(a = "Python",b = "Progrmming", c = "Language")
string
```
